```python
# query_rewriter = to make queries more specific and detailed, improving the likelihood of retrieving.
# step-back_prompting = to generate broader more general queries that can help retrieve erelvant background information..
# sub-query_decomposition = to break down complex queries into simpler sub-queries for more comphrensive information retrieval.
```

```python
re_write_llm  = llm
query_rewrite_template = """You are an AI assistant tasked with reformulating user queries to improve retrieval in a RAG system. 
Given the original query, rewrite it to be more specific, detailed, and likely to retrieve relevant information.

Original query: {original_query}

Rewritten query:    """

query_rewrite_prompt = PromptTemplate(
    input_variables=["original_query"],
    template=query_rewrite_template
)

#creating llm chain for query rewrtiting
query_rewriter = query_rewrite_prompt | re_write_llm
def rewrite_query(original_query):
    response = query_rewriter.invoke(original_query)
    return response.content
```

```python
#testing
original_query = "what are the impacts of climate change on the environment?"

rewrite_query = rewrite_query(original_query)
print("original query: ", original_query)
print("\nrewritten query: ", rewrite_query)
```
